```python
import matplotlib.pylab as plt
%matplotlib inline
import numpy as np
```

```python
# from https://stackoverflow.com/questions/65523844/colormap-diverging-from-black-instead-of-white

from matplotlib.colors import LinearSegmentedColormap
color_min    = "red"
color_center = "gray"
color_max    = "blue"
cmap = LinearSegmentedColormap.from_list(
    "cmap_name",
    [color_min, color_center, color_max]
)
```

```python
x =np.arange(100)
```

```python
plt.scatter(x,x,c=x, cmap=cmap)
plt.colorbar()
```
